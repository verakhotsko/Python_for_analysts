## Задание 3.
Изучите взаимосвязь компаний производителей ноутбуков и компаний производителей процессоров, используя сложенную или многорядовую столбчатую диаграмму
Процессоры от Samsung не изучайте

```python
data_relation = pd.crosstab(columns=data['Cpu_Company'], 
                            index=data['Company']).reset_index()
data_relation.drop(columns='Samsung', inplace=True)
data_relation
```

### 3.1 Постройте график в абсолютных величинах

```python
plt.figure(figsize=(10, 6))
count_ticks = np.arange(len(data_relation['AMD']))
print(count_ticks)
offset=0.2
w=0.4
plt.bar(count_ticks - offset, data_relation['AMD'], width=w)
plt.bar(count_ticks + offset, data_relation['Intel'], width=w)
plt.title('Notebooks')
plt.xlabel('Company')
plt.ylabel('Amount')
plt.xticks(count_ticks, data_relation['Company'], rotation=45)
plt.legend(['AMD', 'Intel'])
```

### 3.2 Постройте график в относительных величинах

```python
relation = pd.crosstab(columns=data['Cpu_Company'], index=data['Company'], normalize='index').reset_index()
relation.drop(columns='Samsung', inplace=True)
relation
```

```python
plt.figure(figsize=(10, 6))
plt.bar(relation['Company'], relation['AMD'] )
plt.bar(relation['Company'], relation['Intel'], bottom=relation['AMD'])

plt.title('Ноутбуки')
plt.xlabel('Производители ноутбуков')
plt.ylabel('Количество')
plt.yticks(np.arange(0,1,0.05),np.arange(0,100,5))
plt.xticks(rotation=45)
plt.legend(['AMD','Intel'])
plt.grid();
```

## Задание 5*.
Постройте матрицу корреляций для таблицы

```python
data_corr = data.corr().round(2)
data_corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data_corr,
            annot=True, 
            linewidths=0.5, 
            cmap='coolwarm', 
            mask=np.triu(data_corr));
```
