```python
import pandas as pd
data = pd.read_csv('laptop_price.csv', index_col='laptop_ID')
data
```

### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
data['Cpu_Company'] = data['Cpu'].apply(lambda x: x.split()[0])
data
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
data['Memory'].value_counts()
```

```python
# 512GB SSD +  1.0TB Hybrid 
def transform_string(row: str) -> int:
    GB = 0
    row_list = row.split()
    for element in row_list:
        if 'GB' in element:
            GB += int(element[: -2])
        elif 'TB' in element:
            GB += float(element[: -2]) * 1024
    return int(GB)       


data['Memory_Amount'] = data['Memory'].apply(transform_string)
data.sample(10)
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
%%time
# 512GB SSD +  1.0TB Hybrid 
def transform_string(row: str) -> str:
    type_ = ''
    list1 = row.split()
    for element in list1:
        if element.isalpha():
            type_ += element + ' '
    return type_

data['Memory_Type'] = data['Memory'].apply(transform_string)
data.sample(5)
```

```python
%%time
def remove_string(row: str) -> str:
    list1 = row.split()
    for i in range(len(list1)-1, -1, -1):
        if 'GB' in list1[i] or 'TB' in list1[i]:
            del list1[i]
    return ' '.join(list1)

data['Memory_Type_test'] = data['Memory'].apply(remove_string)
data.sample(5)
```

### 1.4 Удалите признаки Memory и ScreenResolution

```python
data1 = data.drop(columns=['Memory', 'ScreenResolution'])
data1
```

## Задание 4.
Составьте несколько сводных таблиц

### 4.1 Найдите среднюю стоимость ноутбуков в зависимости от компании производителя
Отсортируйте от меньшей стоимости к большей

```python
mean_price = data.groupby('Company').agg({'Price_euros': 'mean'})
mean_price.sort_values('Price_euros')
```

### 4.2 Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора

```python
mean_price = data.groupby('Cpu_Company').agg({'Price_euros': 
                                      ['min','mean', 'max']})
mean_price
```

### 4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ

```python
data.pivot_table(index='Cpu_Company',
                columns='Ram',
               values='Weight',
               aggfunc='count',
               fill_value=0)
```

### 4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти

```python
data.groupby(['OpSys', 'Memory_Amount'])['Price_euros'].mean()
```

```python
data.groupby(['OpSys', 'Memory_Amount']).agg({'Price_euros': 'mean'}).unstack('Memory_Amount').fillna(0)
```
