## Задание 1.

```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv',encoding='1251', index_col='laptop_ID')
df.sample(10)
```

### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)


```python
df['Cpu'].value_counts()
```

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
df.head()
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)


```python
df['Memory'].value_counts()
```

```python
# 512GB SSD +  1.0TB Hybrid 
def transform(row:str) -> str : 
    gb = 0
    list1 = row.split()
    for el in list1:
        if 'GB' in el:
            gb += int(el[: -2])
        elif 'TB' in el:
            gb += float(el[: -2])*1024
    return int(gb)
    
df['Memory_Amount'] = df['Memory'].apply(transform)
df.sample(5)

```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)


```python
# 512GB SSD +  1.0TB Hybrid 
def transform(row:str) -> str : 
    type_ = ''
    list1 = row.split()
    for el in list1:
        if el.isalpha():
            if type_:
                type_ += ' '
            type_ += el
    return type_

def transform_2(row:str) -> str: 
    list1 = row.split()
    for i in range(len(list1)-1,-1,-1):
        if 'GB' in list1[i] or 'TB' in list1[i]:
            del list1[i]
    return ' '.join(list1)


df['Memory_Type'] = df['Memory'].apply(transform_2)
df.sample(15)
```

### 1.4 Удалите признаки Memory и ScreenResolution

```python
df.drop(columns=['Memory','ScreenResolution'], inplace=True)
df.sample(5)
```

### Задание 4.
Составьте несколько сводных таблиц
####  4.1 Найдите среднюю стоимость ноутбуков в зависимости
от компании производителя
Отсортируйте от меньшей стоимости к большей



```python
df.groupby('Company').agg({'Price_euros':'mean'}).sort_values('Price_euros')
```

#### 4.2 Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора


```python
df.groupby('Cpu_Company').agg({'Price_euros':['min','mean','max']})
```

#### 4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ


```python
df.pivot_table(index='Cpu_Company',
               columns='Ram',
               values='Weight',
               aggfunc='count',
               fill_value=0
                  )
```

#### 4.4 Постройте таблицу с подсчетом средней стоимостиноутбуков в данных в зависимости от операционной системы и GB памяти

```python
# df.groupby(['OpSys','Memory_Amount'])['Price_euros'].mean()
df.groupby(['OpSys','Memory_Amount']).agg({'Price_euros':'mean'}).unstack('Memory_Amount').fillna(0).round(1)
```
