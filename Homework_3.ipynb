```python
import pandas as pd
houses = pd.read_csv('kc_house_data.csv')
houses.head()
```

### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома

```python
houses['delta_renovated'] = houses.apply(lambda x: x['yr_renovated'] - x['yr_built'] if x['yr_renovated'] != 0 else 0, axis=1)
houses.head()
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
houses['date']
```

```python
houses['yr_sale '] = houses['date'].apply(lambda x: x[:4])  
houses.head()
```

```python
houses['month_sale '] = houses['date'].apply(lambda x: x[4:6])  
houses.head()
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
houses_del_columns = houses.drop(columns=['date', 'zipcode', 'lat', 'long'])
houses_del_columns.head()
```

## Задача 3
Составьте несколько сводных таблиц

### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
houses.pivot_table(index='bedrooms',
                  values='price',
                  aggfunc='mean').sort_values('price')
```

```python
houses_group_bedroom = houses.groupby('bedrooms')
houses_group_bedroom.agg({'price': 'mean'}).sort_values('price')
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
houses_gr_condition = houses.groupby('condition')
houses_gr_condition.agg({'price': ['min', 'mean', 'max']})
```

```python
houses.pivot_table(index='condition',
                  values='price',
                  aggfunc=['min', 'mean', 'max'])
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
houses.pivot_table(index=['waterfront', 'view'],
                  values='id',
                  aggfunc='count')
```

```python
houses_group_two = houses.groupby(['waterfront', 'view'])
houses_group_two.agg({'id': 'count'})
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index=houses['floors'], columns=houses['bedrooms'])
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
houses.pivot_table(index='condition', columns='grade', values='price', aggfunc='median', fill_value=0)
```
