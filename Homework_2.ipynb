## Задача 5.

### 5.1 Выберите дом клиенту
Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?

```python
# если запрос "с видом на набережную" соответствует оценке вида 4 и при этом наличие набережной
client_house = house[(house['waterfront'] == True) & (house['view'] == 4) & (house['bathrooms'] >= 3) & (house['sqft_basement'] > 0)]
client_house.shape[0]
```

### 5.2 Выберите дом клиенту
Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?

```python
house1980_good = house[(house['yr_built'] >= 1980) & (house['condition'] > 4)]
client_house = house1980_good[(house1980_good['view'] == 4) | (house1980_good['waterfront'] == True)]
client_house['price'].min(), client_house['price'].max()
```
