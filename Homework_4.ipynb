### 1.1. Скачать следующие данные: kc-house-data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

houses = pd.read_csv('kc_house_data.csv', index_col='id')
houses.head()
```

1. максимальное количество домов (более 5 000) продается в диапазоне цен менее 500 000 евро;

4. максимальная цена продажи в диапазоне от 7 до 8 млн евро

### 2.1. Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
waterfront = houses['waterfront'].value_counts()
waterfront
```

```python
plt.figure(figsize=(8,6))
plt.pie(waterfront, labels=['Нет', 'Есть'], autopct='%1.1f%%')
plt.title('Распределение домов от наличия вида на набережную');
```

#### Вывод:
Данное распределение показывает, что у 99,2% домов имеется набережная. Однако это не говорит нам, есть ли вид из дома на эту набережную.

### Представим, что вид на набережную из дома оценивается на 4 (максимальная оценка) в колонке "view":

```python
not_view_waterfront = houses[((houses['waterfront'] == 1) & (houses['view'] == 3)) | ((houses['waterfront'] == 1) & (houses['view'] == 2)) | ((houses['waterfront'] == 1) & (houses['view'] == 1)) | ((houses['waterfront'] == 1) & (houses['view'] == 0) | (houses['waterfront'] == 0))]
not_v = not_view_waterfront.shape[0]
not_v
```

```python
view_waterfront = houses[(houses['waterfront'] == 1) & (houses['view'] == 4)]
yes_v = view_waterfront.shape[0]
yes_v
```

```python
plt.figure(figsize=(8,6))
plt.pie([not_v, yes_v], labels=['Нет', 'Есть'], autopct='%1.1f%%')
plt.title('Распределение домов от наличия вида на набережную');
```

### Вывод:
Как видим, доля отличного вида на набережную практически не изменилась (уменьшилась на 0,2%).
Однако это более точное распределение исходя из запроса.

## Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

#### Для исследования зависимости цены от характеристик дома, воспользуемся коррелляционной матрицей. Она показывает линейные связи (прямые или обратные) между признаками.

```python
houses_corr = houses.corr().round(1)
houses_corr
```

#### Рассмотрим какие признаки данной выборки входят в какой диапазон корреляции.
#### 1. Сильная корреляция:

```python
strong_corr = houses_corr['price'] >= 0.7
strong_corr
```

#### Итак, сильной корреляцией с ценой обладают признаки "жилая квадратура" и "оценка конструкции здания и дизайн". Т.е. чем больше жилой площади и чем выше оценка, тем выше стоимость здания. 
Давайте рассмотрим подробнее взаимное распределение данных признаков:
### 1.1 Зависимость цены продажи дома ("price") от жилой площади дома ("sqft_living"):

```python
sns.jointplot(x=houses['sqft_living'], y=houses['price']);
```

#### Как видим разброс цен как и жилой квадратуры большой, а плотность выборки сосредоточена в диапазоне до 1 млн. евро и до 3 000 кв.м. Т.е. основные продажи происходили в этих диапазонах.
#### Сократим данную выборку до этих дипазонов:

```python
houses_limit = houses[(houses['price'] < 1000000) & (houses['sqft_living'] < 3000)]
sns.jointplot(x=houses_limit['sqft_living'], y=houses_limit['price'], kind='hex', color='red');
```

#### И посмотрим на коэффициент корреляции в сокращенной выборке:

```python
limit_corr = houses_limit.corr().round(1)
limit_corr['price']['sqft_living']
```

#### Как видим при сокращении выборки до диапазона нормального распределения коэффициент корреляции показывает уже не сильную зависимость, а среднюю (0,5). 
#### Давайте посмотрим относительное количество домов данного диапазона:

```python
houses_limit.shape[0] / houses.shape[0] * 100
```

#### Как видим цена 83 %  домов имеют среднюю зависимость от площади жилых помещений.

#### А что же происходит с оставшейся частью выборки?

```python
houses_limit = houses[(houses['price'] > 1000000) & (houses['sqft_living'] > 3000)]
sns.jointplot(x=houses_limit['sqft_living'], y=houses_limit['price'], kind='hex', color='red');
```

```python
limit_corr = houses_limit.corr().round(1)
limit_corr['price']['sqft_living']
```

#### И здесь коэффициент корреляции упал до среднего значения (0,5).
#### Можно сделать вывод, что данная выборка не совсем подходит для точного анализа этой зависимости и нуждается в дополнительной очистке исходя из задач анализа.
